```
def similarityFunc(listOld,newP):
  avgP = 0
  arrHasil = []
  newP,unwanted = fixNull1(newP)
  avgP = np.mean(newP[1:])
  for i in range(1,len(listOld)-1):
    atas = 0
    bawah = 0
    bawah1 = 0
    bawah2 = 0
    avgElm = np.mean(listOld[i][1:])
    elmList = fixNull2(listOld[i],unwanted)
    for j in range(1,len(elmList)):
      atas += (newP[j]-avgP)*(elmList[j]-avgElm)
      bawah1 += np.square(newP[j]-avgP)
      bawah2 += np.square(elmList[j]-avgElm)
    bawah = np.sqrt(bawah1)*np.sqrt(bawah2)
    hasil = atas/bawah
    arrTemp= []
    arrTemp.append(i)
    arrTemp.append(hasil)
    arrHasil.append(arrTemp)
    arrHasil = sorted(arrHasil, key = lambda x: x[1], reverse=True)[:5]
    
  print("SIMILIAR MATRIX : ",arrHasil)
  return arrHasil
```

```
def fixNull1(person):
  unwanted = [i for i, x in enumerate(person) if x == 0]
  newP = [x for x in person if x != 0]
  return newP,unwanted
```

```
def fixNull2(listOld,unwanted):
  tmplist=listOld
  for i in reversed(unwanted):
    tmplist.pop(i)
  return tmplist
```

```
def predictFunc(newP,arrHasil,listOld):
  newP,unwanted = fixNull1(newP)
  avgP = np.mean(newP[1:])
  down1 = 0
  down21 = 0
  down22 = 0
  down1 = sum(x[1] for x in arrHasil)
  for i in range(len(arrHasil)):
    down21 += arrHasil[i][1] 
    down22 += arrHasil[i][1]*(listOld[arrHasil[i][0]][unwanted[0]] - np.mean(listOld[arrHasil[i][0]][1:]))
  hasilPred = avgP + (down22/(down21))
  return hasilPred
```
